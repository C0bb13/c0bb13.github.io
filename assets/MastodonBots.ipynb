```python
# Bot information comes from when I created the bot account on the Mastodon website, hosted on mastodon.social
# Used to get the access token of the first bot account
def get_yellen_token():
    mastodon_url = 'https://mastodon.social'
    client_id = '5O3CcEJ9f6zMEeIuYGC2ms0tiVh_QN1kA8XQDPzsKho'
    client_secret = 'WVcbPVWNPsgN3TypPQeuOTj506W5UEzjCTKdI6GcFh0'
    username = 'janety121@yahoo.com'
    password = 'vwy!tuTLMG.u67U'

    mastodon = Mastodon(
        client_id=client_id,
        client_secret=client_secret,
        api_base_url=mastodon_url
    )

    access_token = mastodon.log_in(
        username=username,
        password=password,
        scopes=['read', 'write'],
        to_file='access_token.txt'
    )

    return mastodon, access_token

# Used to get the access token of the second bot account
def get_travis_token():
    mastodon_url = 'https://mastodon.social'
    client_id = 'o8RR7FxRTAYuYnd_OYsIW3W9mMtnDqbTV0swIlc85Vo'
    client_secret = 'TA9Wgr3xgQ0lBpRT9nRxavH_rHjn7JbUvON6CYD0l-M'
    username = 'realtraviskelce@yahoo.com'
    password = 'travispassword'

    mastodon = Mastodon(
        client_id=client_id,
        client_secret=client_secret,
        api_base_url=mastodon_url
    )

    access_token = mastodon.log_in(
        username=username,
        password=password,
        scopes=['read', 'write'],
        to_file='access_token.txt'
    )

    return mastodon, access_token

def get_ibn_khaldun_token():
    mastodon_url = 'https://mastodon.social'
    client_id = '_DHWsaOBowG0HhPgw8lTu_xu66kVZQFIb9q53RV3qqk'
    client_secret = '5XBrf6Oo8q4lAICPQgXc1LmnLmakdSzvTffeG2Z7vAs'
    username = 'ibnkhaldun1@yahoo.com'
    password = 'FakeAccount3'

    mastodon = Mastodon(
        client_id=client_id,
        client_secret=client_secret,
        api_base_url=mastodon_url
    )

    access_token = mastodon.log_in(
        username=username,
        password=password,
        scopes=['read', 'write'],
        to_file='access_token.txt'
    )

    return mastodon, access_token
# Used to located the user we want to respond to
def get_user_id(api, username):
    user_info = api.account_search(q=username)
    if user_info and len(user_info) > 0:
        return user_info[0]['id']
    else:
        return None

# Used to get the id of the post itself
def get_most_recent_post(api, user_id):
    user_timeline = api.account_statuses(user_id)
    if user_timeline:
        return user_timeline[0]
    else:
        return None

# Used to generate our unique response!
def generateresponse(content):
    instructions = input("What type of encouragement would you like? ")
    
    # We input our instructions for how we want our bot to respond.
    response = lm.chat(f'''
    System: Provide a {instructions} comment to my prompt.

    User: {content}

    Assistant: 
    ''')

    return response

# This is how I respond to the original post
def post_response(api, original_post_id, response_content):
    api.status_post(
        status=response_content,
        in_reply_to_id=original_post_id
    )

    
# Activate yellen bot
def run_mastodon_bot1():
    # Get Mastodon API instance and access token
    mastodon_api, access_token = get_yellen_token()

    # Ask user for the target username
    target_username = input("Enter the username of the user you want to copy the post from: ")

    # Get user ID
    user_id = get_user_id(mastodon_api, target_username)

    if user_id:
        # Get the most recent post
        recent_post = get_most_recent_post(mastodon_api, user_id)
        if recent_post:
            # Extract and print the content without <p> tags
            post_content = recent_post['content']
            post_content_without_tags = post_content.replace('<p>', '').replace('</p>', '').replace('"', '')

            # Use the most recent post's content in the generateresponse function
            generated_response = generateresponse(post_content_without_tags)
            print("Generated Response:")
            print(generated_response)# Activate Yellen bot

            # Post the generated response as a reply to the original post
            post_response(mastodon_api, recent_post['id'], generated_response)
            print("Response posted successfully!")
        else:
            print(f"No recent posts found for user {target_username}.")
    else:
        print(f"User {target_username} not found.")

        
# Activate travis bot
def run_mastodon_bot2():
    # Get Mastodon API instance and access token
    mastodon_api, access_token = get_travis_token()

    # Ask user for the target username
    target_username = input("Enter the username of the user you want to copy the post from: ")

    # Get user ID
    user_id = get_user_id(mastodon_api, target_username)

    if user_id:
        # Get the most recent post
        recent_post = get_most_recent_post(mastodon_api, user_id)
        if recent_post:
            # Extract and print the content without <p> tags
            post_content = recent_post['content']
            post_content_without_tags = post_content.replace('<p>', '').replace('</p>', '').replace('"', '')

            # Use the most recent post's content in the generateresponse function
            generated_response = generateresponse(post_content_without_tags)
            print("Generated Response:")
            print(generated_response)

            # Post the generated response as a reply to the original post
            post_response(mastodon_api, recent_post['id'], generated_response)
            print("Response posted successfully!")
        else:
            print(f"No recent posts found for user {target_username}.")
    else:
        print(f"User {target_username} not found.")

        
# Activate ibn khaldun bot
def run_mastodon_bot3():
    # Get Mastodon API instance and access token
    mastodon_api, access_token = get_ibn_khaldun_token()

    # Ask user for the target username
    target_username = input("Enter the username of the user you want to copy the post from: ")

    # Get user ID
    user_id = get_user_id(mastodon_api, target_username)

    if user_id:
        # Get the most recent post
        recent_post = get_most_recent_post(mastodon_api, user_id)
        if recent_post:
            # Extract and print the content without <p> tags
            post_content = recent_post['content']
            post_content_without_tags = post_content.replace('<p>', '').replace('</p>', '').replace('"', '')

            # Use the most recent post's content in the generateresponse function
            generated_response = generateresponse(post_content_without_tags)
            print("Generated Response:")
            print(generated_response)

            # Post the generated response as a reply to the original post
            post_response(mastodon_api, recent_post['id'], generated_response)
            print("Response posted successfully!")
        else:
            print(f"No recent posts found for user {target_username}.")
    else:
        print(f"User {target_username} not found.")
```
